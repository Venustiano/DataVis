```r
library(ggplot2)
```

```r
tableRes[, ("cat") := 'participants']
```

```r
# Basic violin plot
p <- ggplot(tableRes, aes(x= cat,y=medSpeed)) + 
  geom_violin()
p
```

```r
# Long format
longTable <- melt(tableRes,id.vars=c("Age","iSubj","trial","Decade","cat"),
                      variable.name="BalanceMeasures",
                      value.name="value")
longTable
```

```r
p <- ggplot(longTable, aes(x= BalanceMeasures,y=value)) + 
  geom_violin()
p
```

```r
# The most common normalization is the z-transformation, where you subtract 
# the mean and divide by the standard deviation of your variable. 
# The result will have mean=0 and sd=1.
longTable[,("normValues"):=(value - mean(value)) / sd(value),by = (BalanceMeasures)]
str(longTable)
```

```r
options(repr.plot.width=12, repr.plot.height=7)
p <- ggplot(longTable, aes(x= BalanceMeasures,y=normValues)) + 
  geom_violin()
p
```

```r
longTable[,grupo:=ifelse(Age>60,2,1)]
str(longTable)
```

```r
options(repr.plot.width=14, repr.plot.height=7)
p <- ggplot(longTable, aes(x= BalanceMeasures,y=normValues,color=factor(grupo))) + 
  geom_violin(alpha=0.3,position="identity")
p
```

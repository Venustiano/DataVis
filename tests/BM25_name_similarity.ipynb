### 2.9 Tokenize lists for BM25 algorithm

```python
l_names =list(names[0].str.strip().unique())
print("Names:",len(l_names))
```

```python
# we may improve results if we make lists of male and female names
tokenized_names = [[char for char in name] for name in l_names]

tokenized_names
```

```python
bm25_names = BM25Okapi(tokenized_names)
```

Test to find similar  names

```python
name = 'Mari'
tokenized_name = [char for char in name]
#tokenized_name
top_fnames = bm25_names.get_top_n(tokenized_name, l_names, n=10)
print("Top names:",top_fnames)
```
